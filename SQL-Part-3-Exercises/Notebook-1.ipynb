## **QUESTION 1:**  Exploring a LEFT JOIN using the **books** and **to\_read** tables

## **QUESTION 2:** Exploring **RIGHT** joins with the **books** and **to\_read** tables

You will need to uncomment the Query to answer the question.

**Part A:** Looking at the query, which table is the 'Left' and which is the 'Right'?

> Your Answer: books is the left table and to\_read is the right table.

**Part B:** How do the RIGHT-joined tables differ from the LEFT-joined tables in Question 1?  

> Your Answer: In a Right joined tables all the rows from the right tables are included in the output and only matching rows from the left tables are included. The non-matching rows from the left table will contain null values in the column of right tables.
> 
> In a Left joined tables all the rows from the left tables are included in the output and only matching rows from the right tables are included. The non-matching rows from the right table will contain null values in the column of left tables.

```sql

-- Query 1: RIGHT JOIN
SELECT TOP 300 b.title, b.average_rating, b.books_count, tr.user_id, tr.book_id
FROM BooksDB.dbo.books AS b
RIGHT JOIN BooksDB.dbo.to_read AS tr
ON b.best_book_id = tr.book_id;
```

> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; background-color: rgba(127, 127, 127, 0.1);"><i>Minimum Desired Output:</i>&nbsp;The tag name for the most popular tag used for the most tagged book.&nbsp;</span> 
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; background-color: rgba(127, 127, 127, 0.1);"><b>Tag Name:</b></span>

```sql
-- Code Here:
SELECT b.title, count(bt.tag_id) as tag_count
FROM BooksDB.dbo.books AS b
INNER JOIN BooksDB.dbo.book_tags AS bt
ON b.best_book_id = bt.goodreads_book_id
group by b.title, bt.tag_id
order by tag_count DESC;




SELECT COUNT(t.goodreads_book_id) AS 'Book Tag Count', b.title, t.tag_id
FROM BooksDB.dbo.book_tags AS t
INNER JOIN BooksDB.dbo.books AS b
ON b.best_book_id = t.goodreads_book_id
GROUP BY b.title, t.tag_id
ORDER BY 'Book Tag Count' DESC;


SELECT b.title, count(bt.tag_id) as t
from BooksDB.dbo.books as b
join BooksDB.dbo.book_tags as bt
on b.book_id = bt.goodreads_book_id
group by b.title
--having popular(tag_id)
order by t DESC

```
